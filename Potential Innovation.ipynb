```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the Data
spotify = pd.read_csv('/Users/antineutrino/Desktop/Team_4_Assignment_1.4/spotify_tracks-1.csv')

# Display a few lines from the dataframe
spotify.head()
```

```python
# A new dataframe where the popularity is no less than 90
popular_spotify = spotify[spotify['popularity'] >= 90]

# Display the first few rows of the new dataframe to verify
popular_spotify
```

```python
# Filter tracks based on the popularity thresholds
above_80 = spotify[spotify['popularity'] > 80]
above_85 = spotify[spotify['popularity'] > 85]
above_90 = spotify[spotify['popularity'] > 90]

# Ensure at least 20 tracks with popularity > 80
selected_above_80 = above_80.sample(n=20)

# Ensure at least 10 tracks with popularity > 85
remaining_above_85 = above_85.drop(selected_above_80.index, errors='ignore')
selected_above_85 = remaining_above_85.sample(n=10)

# Ensure at least 5 tracks with popularity > 90
remaining_above_90 = above_90.drop(selected_above_80.index, errors='ignore')
selected_above_90 = remaining_above_90.sample(n=5)

# Combine the selections
random_tracks = pd.concat([selected_above_80, selected_above_85, selected_above_90]).drop_duplicates()

# If the combined selections are less than 30, randomly select remaining tracks from the dataset
if len(random_tracks) < 30:
    remaining_tracks_needed = 30 - len(random_tracks)
    remaining_tracks = spotify.drop(random_tracks.index, errors='ignore')
    additional_tracks = remaining_tracks.sample(n=remaining_tracks_needed)
    random_tracks = pd.concat([random_tracks, additional_tracks])

# Display the selected tracks
random_tracks.head()
```

```python
# Create a list of tracks that are in popular_spotify but not in random_tracks
recommend_tracks = popular_spotify[~popular_spotify['track_id'].isin(random_tracks['track_id'])]

# Display the recommended tracks
recommend_tracks.head()
```
